```python
from stoneforge.io.lasparser import parse_las_file

filepath = "../stoneforge/datasets/example_las3.las"
data = parse_las_file(filepath)
data.keys()
```

```python
class LASaccess:
    def __init__(self, filepath, version=3):
        self.filepath = filepath
        if version == 3:
            self.data = parse_las_file(filepath)
            self.tables = list(self.data.keys())
            
        if version == 2:
            raise NotImplementedError("LAS version 2 is not supported yet.")

    def get_data(self):
        return self.data
```

```python
las_manager = LASaccess(filepath, version=3)
las_data = las_manager.get_data()
```
