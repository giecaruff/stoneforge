```python
from stoneforge.data_management.preprocessing import DataLoader
```

```python
# Update DLIS object with the path to your DLIS file

dlis_obj = DataLoader(r"DSDP_leg_96_hole_616_96_processed_data.dlis", filetype = 'dlis')
```

```python
# Data Access Preview

dlis_obj.data_obj.show_header()
```

```python
# All mnemonics:

mns = dlis_obj.data_obj.mnemonics()
print(mns)
```

```python
# Import and display data structure

data = dlis_obj.data_obj.extract(mnemonics=mns)
for d in data:
    print("digital files:",d, "| frames:",data[d].keys())
```
