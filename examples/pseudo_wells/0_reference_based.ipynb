```python
from stoneforge.pseudo_wells import anadrill_siliciclastic
import matplotlib.pyplot as plt
```

```python
structure = {
    1: 4000,   # shale
    3: 2000,   # oil sand
    8: 2050,   # gas sand
    2: 1950   # carbonate
}

ee = anadrill_siliciclastic(structure, top = 1000.0, bottom = 2000.0, step=0.25, random_state=False)

plt.plot(ee['DEPTH'], ee['NPHI'])
plt.grid()
plt.show()
```
