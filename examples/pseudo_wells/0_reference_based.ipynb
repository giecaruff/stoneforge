```python
# Importing necessary libraries

from stoneforge.pseudo_wells import anadrill_siliciclastic
from stoneforge.vis import wellplot, plito

import matplotlib.pyplot as plt
import pandas as pd
```

```python
anadrill_siliciclastic()
```

```python
# Define facies and sampling structure
# For 'anadrill_siliciclastic' the facies are (in order):
# 0: shale; 1: clean_sandstone with gas; 2: clean_sandstone with oil; 3: clean_sandstone with brine; 4: feldspatic sandstone
# 5: unconsolidated sandstone with fresh water; 6: organic shale; 7: siltite; 8: dirty sandstone with brine

structure = (
    [0, 1, 2, 3, 4, 0, 5, 0, 6, 7, 8, 3, 0],  # Facies list
    [400]*13  # Counts list
)

well,units = anadrill_siliciclastic(structure, top = 1000.0, bottom = 2000.0, step=0.25, random_state=False)
```

```python
well.keys()
```

```python
units
```

```python
data = pd.DataFrame.from_dict(well)
data
```

```python
units
```

```python
curves = ['GR','NPHI', 'DEN', 'DT', 'CODE']
units2 = []
for c in curves:
    units2.append(units[c])

color = ["green","navy","red","magenta","black"]

wellplot(data,"DEPTH",curves, color, units2)
```

```python
colors ={
    22:"darkgreen",
    27:"orange",
    25:"grey",
    49:"gold",
    48:"yellow",
    57:"green",
    54:"maroon",
    0:"black"
}

plt.subplot(1, 3, 1)
plt.plot(data["GR"],data["DEPTH"],color="green")
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(data["NPHI"],data["DEPTH"],color="navy")
plt.grid()

plt.subplot(1, 3, 3)
plito(data["CODE"],data["DEPTH"],colors)
plt.grid()
```
