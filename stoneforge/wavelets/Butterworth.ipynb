```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import pylops

imp = signal.unit_impulse(240, 'mid')
b, a = signal.butter(4, 55,fs=240)
response_zp = signal.filtfilt(b, a, imp)
```

```python
plt.plot(np.arange(-120, 120),response_zp)
plt.margins(0.1, 0.1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
spec = abs(np.fft.rfft(response_zp))
plt.plot(spec)
plt.xlabel('FrequÃªncia (hz)')
plt.ylabel('Amplitude relativa')
plt.grid()
plt.show()
```

```python
low_b, low_a = signal.butter(2,1.5,'hp', fs=240)
Low_cut_response_zp = signal.filtfilt(low_b, low_a, response_zp)
```

```python
plt.plot(np.arange(-120, 120),Low_cut_response_zp)
plt.xlim(-20,20)
plt.margins(0.1, 0.1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
